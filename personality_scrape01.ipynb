```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
from datetime import datetime
import pandas as pd
import requests
from bs4 import BeautifulSoup
import json
```

```python
# Url being scraped
url = "https://www.personality-database.com/profile?pid=1"

# Using requests.get to scrape the url
response = requests.get(url)

# Using BeautifulSoup to get the html from the url
soup = BeautifulSoup(response.text)

#Printing the soup
print(soup)
```

```python
celeb_types = soup.find_all('div',class_='profile-card')
celeb_types
```
